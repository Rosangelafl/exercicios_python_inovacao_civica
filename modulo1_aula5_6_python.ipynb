```python
import glob
```

```python
dados_todos_estados = glob.glob('/content/drive/MyDrive/Python para inovação civica/semana_1/base_dados/*.csv')
print(dados_todos_estados)
```

```python
lista_df= []
for dados_estado in dados_todos_estados:
	df_estado = pd.read_csv(dados_estado, sep=';', encoding='Latin-1')
	lista_df.append(df_estado) 
```

```python
lista_df
```

```python
candidaturas_pe_rn = pd.concat(lista_df, axis=0, ignore_index=True)
```
