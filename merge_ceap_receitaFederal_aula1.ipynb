```python
import pandas as pd
```

```python
DTYPE = {'cnpj': str,}
```

```python
df_empresas = pd.read_csv('/content/drive/MyDrive/Python para inovação civica/semana_4/2019-11-19-companies.csv.xz', 
    compression='xz', dtype=DTYPE, parse_dates=['situation_date'])
```

```python
df_empresas.head(3)
```

```python
df_empresas["cnpj"] = df_empresas["cnpj"].str.replace(r'\D', '', regex=True)
```

```python
df_empresas.situation.unique()
```

```python
tipos = {
    'txNomeParlamentar': str,
    'ideCadastro': str,
    'nuCarteiraParlamentar': str,
    'nuLegislatura': str,
    'sgUF': str,
    'sgPartido': str,
    'codLegislatura': str,
    'numSubCota': str,
    'txtDescricao': str,
    'numEspecificacaoSubCota': str,
    'txtDescricaoEspecificacao': str,
    'txtFornecedor': str,
    'txtCNPJCPF': str,
    'txtNumero': str,
    'indTipoDocumento': str,
    'datEmissao': str,
    'vlrDocumento': float,
    'vlrGlosa': str,
    'vlrLiquido': float,
    'numMes': str,
    'numAno': str,
    'numParcela': str,
    'txtPassageiro': str,
    'txtTrecho': str,
    'numLote': str,
    'numRessarcimento': str,
    'nuDeputadoId': str,
    'ideDocumento': str,
}
```

```python
df_ceap = pd.read_csv('/content/drive/MyDrive/Python para inovação civica/semana_3/Ano-2019.csv', sep=';',
                      encoding='utf-8', error_bad_lines=False, low_memory=True,
                      parse_dates=['datEmissao'], dtype=tipos).copy()
```

```python
df_ceap.head(2)
```

```python
df_ceap.columns.values
```

```python
 df_ceap['txtCNPJCPF'] = df_ceap['txtCNPJCPF'].astype('string')
```

```python
# vamos trocar qualquer caracter que não seja um dígito número por '', essa operação remove os dígitos
df_ceap['txtCNPJCPF'] = df_ceap['txtCNPJCPF'].str.replace(r'\D', '', regex=True)
```

```python
df_final = df_ceap.merge(df_empresas, how='left',
                         left_on='txtCNPJCPF', right_on='cnpj')
```
